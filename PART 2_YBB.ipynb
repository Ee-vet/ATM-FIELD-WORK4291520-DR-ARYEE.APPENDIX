```python
# importing required libraries
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# DATA ASSIMILATION WITH OPTIMAL INTERPOLATION

```python
#loading observation data( in degree celcius)
o_d= np.array([15.2, 16.1, 14.5, 15.8, 25.0 ])

#loading model_output data( in degree celcius)
m_od= np.array([14.8, 15.5, 14.0, 16.0, 25.3 ])

                                #- PERFORMING OPTIMAL INTERPOLATION-
#Calculating weights of both dataset
# calculating variances
var_ob = np.var(o_d)
var_mo = np.var(m_od)
# Calculating for required weights
W_ob = 1 / var_ob / (1 / var_ob + 1 / var_mo)
W_mo = 1/ var_mo / (1 / var_ob + 1 / var_mo)

# Calculating Optimal Interpolation
analyzeddata = W_ob*o_d + W_mo*m_od

# Calculating root mean squared error(rsme)
rmse = np.sqrt(np.mean((analyzeddata-o_d)**2))

# Calculating bias error
bias = np.mean(analyzeddata-o_d)

# Printing results
print("Observation_data: ", o_d)
print("Model_output_data: ", m_od)
print("Weight_observed: ", W_ob)
print("Weight_model: ", W_mo)
print("Analyzed_data: ", analyzeddata)
print("RMSE: ", rmse)
print("Bias error: ", bias)
```

# PLOTTING COMPARISION DATASET
